```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Domino del la función de 2 VI
X = np.linspace(-6, 6,90)
Y = np.linspace(-6, 6,90)
X, Y = np.meshgrid(X, Y)

# Variable dependiente V.D
Z=X**2 + Y**2
```

<h2 style="color:#7FA00A">Curvas de nivel o contorno </h2>   

```python

fig,ax=plt.subplots()

#Generar curvas de nivel o contorno
cp = ax.contourf(X, Y, Z) # contourf = dibujar curvas de nivel y rellenar los contornos 

#fig.colorbar(cp) # superficie
ax.set_title('Gráfica curvas de nivel o contorno sin Colorbar')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')

plt.show()
```

<h1 style="color:#7FA00A"> Importancia Colorbar </h1>   


```python

# Configuración general de la figura de la gráfica
fig = plt.figure(figsize=(9,4.2), layout='constrained')
fig.suptitle('¿Dónde está el máximo y el mínimo? para cuando el Colorbar')


# Primera gráfica
ax = fig.add_subplot(1, 2, 1) #número_filas, número_columnas, index

cp = ax.contourf(X, Y, Z) 
ax.set_title('Primera gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
#fig.colorbar(p_surf, shrink=0.5, aspect=9)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 



# Segunda gráfica
ax = fig.add_subplot(1, 2, 2)#número_filas, número_columnas, index

cp = ax.contourf(X, Y, -Z) 
ax.set_title('Segunda gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
#fig.colorbar(cp, shrink=0.5, aspect=9)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 

plt.show()
```

<h2 style="color:#7FA00A"> Activando Colorbar de las curvas de nivel o contorno </h2>   


```python
fig,ax=plt.subplots()

#Generar curvas de nivel o contorno
cp = ax.contourf(X, Y, Z) # contourf = dibujar curvas de nivel y rellenar los contornos 

fig.colorbar(cp) # superficie
ax.set_title('Gráfica curvas de nivel o contorno con Colorbar')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')

plt.show()
```

```python
# Configuración general de la figura de la gráfica
fig = plt.figure(figsize=(9,4.2), layout='constrained')
fig.suptitle('¿Dónde está el máximo y el mínimo? por fin Colorbar')


# Primera gráfica
ax = fig.add_subplot(1, 2, 1) #número_filas, número_columnas, index

cp = ax.contourf(X, Y, Z) 
ax.set_title('Primera gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
fig.colorbar(cp)#superficie



# Segunda gráfica
ax = fig.add_subplot(1, 2, 2)#número_filas, número_columnas, index

cp = ax.contourf(X, Y, -Z) 
ax.set_title('Segunda gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
fig.colorbar(cp)#superficie,

plt.show()
```

<h2 style="color:#7FA00A"> Visualizando gráficas 3D y sus curvas de nivel o contorno </h2>   


```python
# Configuración general de la figura de la gráfica
fig = plt.figure(figsize=(10, 6.6), layout='constrained')
fig.suptitle('Gráficas 3D y sus curvas de nivel o contorno con Colorbar')


# Primera gráfica
ax = fig.add_subplot(2, 2, 1, projection='3d') #número_filas, número_columnas, index

p_surf = ax.plot_surface(X, Y, Z,cmap = plt.cm.viridis)# cm = Colormap 
ax.set_title(' Primera gráfica')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')


# Segunda gráfica
ax = fig.add_subplot(2, 2, 2)#número_filas, número_columnas, index

cp = ax.contourf(X, Y, Z) 
ax.set_title('Primera gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
fig.colorbar(cp)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 


# Tercer gráfica
ax = fig.add_subplot(2, 2, 3, projection='3d')#número_filas, número_columnas, index

s_surf = ax.plot_surface(X, Y, -Z, cmap = plt.cm.viridis)# cm = Colormap 
ax.set_title(' Segunda gráfica')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')



# Cuarta gráfica
ax = fig.add_subplot(2, 2, 4)#número_filas, número_columnas, index

cp = ax.contourf(X, Y, -Z) 
ax.set_title('Segunda gráfica curvas de nivel o contorno')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')
fig.colorbar(cp)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 

plt.show()
```
