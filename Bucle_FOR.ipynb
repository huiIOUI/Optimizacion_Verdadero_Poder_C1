### Este bucle solo se repite un nùmero fijo de veces.

```python
for iterador_valor in range(1,9): # range(valor_incial, valor_final) finaliza valor_final-1
    print('valor = ',iterador_valor) # Bloque del còdigo que se repite
```

```python
tabla=9

print('Tabla del {}'.format(tabla)) # No está dentro del for por lo tanto se ejecuta o repite una sola vez

for valor in range(1,333): # range(valor_incial, valor_final) finaliza valor_final-1
    resultado=tabla*valor
    print('{}*{} = {}'.format(tabla,valor,resultado))
```
