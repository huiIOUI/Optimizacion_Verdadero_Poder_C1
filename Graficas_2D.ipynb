```python
# Generar valores dominio y rango
x = np.linspace(-6, 6, 100)
y = x**2+3
```

```python
print(x)
```

```python
print(y)
```

```python
# Gràfica
fig, ax = plt.subplots()

ax.plot(x, y)
plt.show()
```

<h1 style="color:#3498DB"> ¿Podemos mejorar la parte visual de la gráfica?</h1>  

```python
# Gràfica
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,linewidth=3, c='orange', label='x**2+3')

ax.set_title('Mejorando parte visual de la gráfica') # Añadir título
ax.set_xlabel('Variable Independiente (V.I) =  x') # Añadir label eje de las abscisas
ax.set_ylabel('Variable dependiente  (V.D) = y') # Añadir label eje de las ordenadas
ax.legend() # Para activar la etiqueta (label)

plt.show()
```

<h1 style="color:#3498DB"> Graficando ecuación de la recta pendiente</h1>  

```python
# Generando valores
x = np.linspace(-3, 3, 100)
y = 2*x+3

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,linewidth=3, c='green', label='2*x+3')

ax.set_title('Gráfica ecuación de la recta pendiente')
ax.set_xlabel('Variable Independiente (V.I) =  x')
ax.set_ylabel('Variable dependiente  (V.D) = y')
ax.legend() # Para activar la etiqueta (label)


ax.axhline(3, color = "orange") 
ax.axvline(0, color = "orange") 

plt.show()
```

```python
# Generando valores
x = np.linspace(-3, 3, 100)
y = 2*x+3

fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,linewidth=3, c='green', label='2*x+3')

ax.set_title('Gráfica ecuación de la recta pendiente')
ax.set_xlabel('Variable Independiente (V.I) =  x')
ax.set_ylabel('Variable dependiente  (V.D) = y')
ax.legend() # Para activar la etiqueta (label)


ax.axhline(3, color = "orange",xmax=0.5) 
ax.axvline(0, color = "orange",ymax=0.5) 

plt.show()
```
