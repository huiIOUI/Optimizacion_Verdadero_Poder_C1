```python
import numpy as np
import matplotlib.pyplot as plt
```

<h2 style="color:#7FA00A"> Entendiendo meshgrid </h2º>   


```python
x = np.linspace(1,3,3)
y = np.linspace(1,6,6)
print('Valores de x {}\nValores de y {}'.format(x,y))
```

```python
x, y = np.meshgrid(x, y)
print('Valores de x {}\nValores de y {}'.format(x,y))
```

```python
# Gràfica
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(x, y,marker='o', color='blue', linestyle='none')

ax.set_title('Entendiendo meshgrid') # Añadir título
ax.set_xlabel('Variable Independiente (V.I) =  x') # Añadir label eje de las abscisas
ax.set_ylabel('Variable dependiente  (V.D) = y') # Añadir label eje de las ordenadas
ax.set_xticks([1,2,3])
plt.show()
```

<h2 style="color:#7FA00A"> Como generar un sistema coordenado rectangular 3D </h2>   


```python
fig = plt.figure(figsize = (3,3))
ax = plt.axes(projection='3d') # Generar sistema coordenado rectangular 3D
plt.show()
```

```python

# Domino del la función de 2 VI
X = np.linspace(-3, 3,90)
Y = np.linspace(-3, 3,90)
X, Y = np.meshgrid(X, Y)

# Variable dependiente V.D
Z= 5*X*Y/np.exp(X**2 + Y**2)
```

```python

fig = plt.figure(figsize = (6,6))
ax = plt.axes(projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cool) # cm = Colormap cool,autumn,summer
#https://matplotlib.org/stable/gallery/color/colormap_reference.html

# Labels configuración
ax.set_title(' Gráfica sistema coordenado rectangular 3D')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')

# Visualización Colorbar
#fig.colorbar(surf, shrink=0.5, aspect=9)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 

plt.show()
```


<h2 style="color:#7FA00A"> Aprendamos algo nuevo </h2>   


```python

# Configuración general de la figura de la gráfica
fig = plt.figure(figsize=plt.figaspect(0.41)) # Factor del aspecto de la figura de la gráfica
fig.suptitle('Dos subplots nivel deidad =)')


# Primera gráfica
ax = fig.add_subplot(1, 2, 1, projection='3d') #número_filas, número_columnas, index

# Domino del la función de 2 VI
X = np.linspace(-3, 3,90)
Y = np.linspace(-3, 3,90)
X, Y = np.meshgrid(X, Y)

# Variable dependiente V.D
Z= 5*X*Y/np.exp(X**2 + Y**2)

p_surf = ax.plot_surface(X, Y, Z,cmap = plt.cm.viridis)# cm = Colormap 

ax.set_title(' Primera gráfica con Colorbar')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')

fig.colorbar(p_surf, shrink=0.5, aspect=9)#superficie,escalar colorbar por un factor,relaciòn dimensiones largo-corto 



# Segunda gráfica
ax = fig.add_subplot(1, 2, 2, projection='3d')#número_filas, número_columnas, index

                 

#  Variable dependiente V.D
s_surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cool)# cm = Colormap 

ax.set_title(' Segunda gráfica sin Colorbar')
ax.set_xlabel('Primera V.I = x')
ax.set_ylabel('Segunda V.I = y')


plt.show()
```
