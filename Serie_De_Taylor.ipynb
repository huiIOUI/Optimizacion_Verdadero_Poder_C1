```python
import sympy as sym
from IPython.display import display #Nos permite visualizar de una forma simb칩lica
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('_mpl-gallery') # A침ade estilo(cuadricula) a las gr치ficas 2D
```

```python
x= sym.symbols('x')
expr_1= sym.sin(x)
display(expr_1)
```

```python
a = 1
taylor= expr_1.subs(x,a)
display(taylor)

grado=6 # El grado real es  grado-1 !

for x1 in range(1,grado+1):
    derivada=sym.diff(expr_1,x,x1)
    taylor+=derivada.subs(x,a)*(x-a)**x1/sym.factorial(x1)
    #display(taylor)
```

```python
display(taylor)
```

```python
values = np.linspace(-3, 3, 150)
u1=sym.lambdify(x,expr_1,'numpy')
u2=sym.lambdify(x,taylor,'numpy')
```

```python
u1(1)
```

```python
# Gr맍ica
fig, ax = plt.subplots(figsize = (5,3))

ax.plot(values, u1(values),linewidth=3, c='orange',label='Original')#label='(x1-1.5)**2+0.2'


ax.plot(values, u2(values),label='Taylor Serie')#label='(x1-1.5)**2+0.2'

ax.scatter(1,u1(1),s=60, color='green',label='valor de referencia')

ax.set_title('Poderosa serie de Taylor grado '+str(grado)) # A침adir t칤tulo
ax.set_xlabel('Variable Independiente (V.I) =  x1') # A침adir label eje de las abscisas
ax.set_ylabel('Variable dependiente  (V.D) = y1') # A침adir label eje de las ordenadas


ax.legend() # Para activar la etiqueta (label)

plt.show()
```

```python
def taylor(grado):
    a = 1
    taylor= expr_1.subs(x,a)
    #display(taylor)

    for x1 in range(1,grado):
        derivada=sym.diff(expr_1,x,x1)
        taylor+=derivada.subs(x,a)*(x-a)**x1/sym.factorial(x1)
    return taylor
```

```python
taylor(15) # El grado real es  grado-1 !
```

```python
plt.style.use('dark_background')
```

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 7.2), layout='constrained') # layout='constrained' Ajusta autom치ticamente subplots y las decoraciones, como legends,colorbars..
grado_ini=2 # Grado inicial, El grado real es  grado_ini-1 !

for ax in axs.flat:
    u3=sym.lambdify(x,taylor(grado_ini),'numpy')
    ax.plot(values, u1(values),color='purple',linewidth=3,label='Original') #x9, u2(x9)
    ax.plot(values, u3(values),color='cyan',label='Taylor '+str(grado_ini-1)) #x9, u2(x9)
    ax.scatter(1,u1(1),s=60, color='green')
    ax.set_title('Taylor grado '+str(grado_ini-1))
    ax.legend()
    grado_ini+=1
```
