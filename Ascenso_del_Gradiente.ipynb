```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec # Nos permite poner subplots dentro de la figura
import numpy as np
```

```python
plt.style.use('_mpl-gallery')
```

<h2 style="color:#3498DB"> Ascenso del gradiente</h2>  

$$F(x) = x^{2/3}(x^2-8)$$

$$ \frac{\mathrm{d}}{\mathrm{d}x}F(x) = \frac{8(x^2-2)}{3x^{1/3}} $$


```python
def funcion_x(x):
    return np.cbrt(x**2)*(x**2-8) #np.cbrt() Calcular la raíz cúbica   
```

```python
def derivada_x(x):
    return (8*(x**2-2))/(3*np.cbrt(x)) #np.cbrt() Calcular la raíz cúbica   
```

<h2 style="color:#3498DB"> Funciòn ascenso del gradiente</h2>  

```python

def ascenso_gradiente(v_vi,alpha,k):   
    '''
    v_vi = valor inicial o de referencia   1.92
    alpha =  learning rate
    k = número de iteraciones
    '''
    vi_x,va_d=np.empty(k),np.empty(k)# almacenar valores V.I y derivada
    for _ in range(0,vi_x.shape[0]):

        derivada=derivada_x(v_vi) # evaluación de la derivada
        
        vi_x[_]=v_vi # almacena valores de la V.I
        va_d[_]=derivada # almacena valores de la derivada

        #print(str(_)+' Variable Independiente',v_vi)
        #print(str(_)+' Derivada ',derivada_x(v_vi),'\n')

        v_vi=v_vi+alpha*derivada # ascenso del gradiente
        
    return vi_x,va_d
    
```

<h2 style="color:#3498DB"> Gráfica del sistema</h2>  

```python
def grafica(v_vi,va_de,log_sc):
    
    '''
    v_vi   = valores de la V.I
    va_de  = valores de la derivada
    log_sc = activar escala logarítmica (eje de las ordenadas)
    '''  
    
    x = np.linspace(-4, 4, 600)
    y = np.cbrt(x**2)*(x**2-8) # np.cbrt() Calcular la raíz cúbica

 
    fig = plt.figure(tight_layout=True,figsize = (9,6))
    fig.suptitle('Ascenso del gradiente OPT =)')
    gs = gridspec.GridSpec(2, 2) # GridSpec(Fila, Columna) Nos permite poner subplots dentro de la figura


    
    # Primera gráfica posición 0,0
    ax = fig.add_subplot(gs[0,0])
    ax.plot(v_vi, color='magenta',linewidth=3)
    ax.set_title('V.I  vs Iteraciòn ')
    ax.set_ylabel('Valores V.I =  x')
    ax.set_xlabel('Iteraciòn')

        
    # Segunda gráfica posición 0,1
    ax = fig.add_subplot(gs[0,1])
    ax.plot(va_de, linewidth=2.1, c='green')
    ax.set_title('Derivada vs Iteraciòn ')
    ax.set_ylabel('Derivada')
    ax.set_xlabel('Iteraciòn')
    if log_sc:
        # Activar escala logarítmica (eje de las ordenadas) 
        ax.set_yscale('symlog') # symlog logaritmo simétrico, permite valores positivos y negativos

    
    
    # Tercer y cuarta gráfica se vuelven solo una, por eso se implementó gridspec.GridSpec() 
    ax = fig.add_subplot(gs[1, :])
    
    # Gráfica función original
    ax.plot(x, y, linewidth=2.1, c='orange')
    
    # Gráfica de valor inicial (punto)
    ax.scatter(v_vi[0],funcion_x(v_vi[0]),c='cyan',s=60, label='Valor inicial = '+str(v_vi[0]))

    # Gráfica valores de la V.I durante las iteraciones
    ax.scatter(v_vi[1:],funcion_x(v_vi[1:]),c='red',s=18)
    ax.set_title('Funciòn original')
    ax.set_xlabel('Variable Independiente (V.I) =  x')
    ax.set_ylabel('Variable dependiente (V.D) =  F(X)')
    ax.legend()

    plt.show()    
```
